### Visualizations on covid19 for Indian states to analyse number of confirmed cases, recovered cases and deceased cases.

### Visualizations for Date and Total number of cases in India:

```python
x=data['Date_YMD']
y=data['TT']
plt.figure(figsize=(15,12))
plt.plot(x,y,color='salmon')
plt.title('Date and total cases of covid19',fontsize=18)
plt.xlabel('Date',fontsize=14)
plt.ylabel('Number of cases',fontsize=14)
plt.show()
```

```python
x=data['Date_YMD']
y=data['TT']
plt.figure(figsize=(15,12))
plt.title('Date and total cases of covid19',fontsize=18)
plt.xlabel('Date',fontsize=14)
plt.ylabel('Number of cases',fontsize=14)
plt.bar(x,y)
plt.show()
```

```python
x=data['Date_YMD']
y=data['TT']
plt.figure(figsize=(18,12))
plt.title('Date and total cases of covid19',fontsize=18)
plt.xlabel('Date',fontsize=14)
plt.ylabel('Number of cases',fontsize=14)
plt.stem(x,y)
plt.show()
```
